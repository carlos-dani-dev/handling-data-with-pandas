#### Dia 2: mapeamento do código de localização do material no acervo para posterior categorização relacionada ao material

```python
# 000 a 099: Generalidades. Ciência e conhecimento.
# 100 a 199: Filosofia e psicologia.
# 200 a 299: Religião.
# 300 a 399: Ciências sociais.
# 400 a 499: Classe vaga. Provisoriamente não ocupada.
# 500 a 599: Matemática e ciências naturais.
# 600 a 699: Ciências aplicadas.
# 700 a 799: Belas artes.
# 800 a 899: Linguagem. Língua. Linguística.
# 900 a 999: Geografia. Biografia. História
def map_loc_class(loc):
    if loc >= 0 and loc <= 99:
        return 'Generalidades. Ciência e conhecimento.'
    if loc >= 100 and loc <= 199:
        return 'Filosofia e psicologia.'
    if loc >= 200 and loc <= 299:
        return 'Religião.'
    if loc >= 300 and loc <= 399:
        return 'Ciências sociais.'
    if loc >= 400 and loc <= 499:
        return 'Classe vaga. Provisoriamente não ocupada.'
    if loc >= 500 and loc <= 599:
        return 'Matemática e ciências naturais.'
    if loc >= 600 and loc <= 699:
        return 'Ciências aplicadas.'
    if loc >= 700 and loc <= 799:
        return 'Belas artes.'
    if loc >= 800 and loc <= 899:
        return 'Linguagem. Língua. Linguística.'
    if loc >= 900 and loc <= 999:
        return 'Geografia. Biografia. História'
```

```python
df_mesclado['classificacao'] = df_mesclado['localizacao'].apply(map_loc_class)
df_mesclado
```

```python
colecao = 'Acervo Circulante'
```

```python
# Avaliar dentre os alunos de graduação e pós graduação a distribuição de empréstimos mensais por ano realizados entre 2010 e 2020 da coleção que tiver a maior frequência de empréstimos.
# Plote um gráfico para cada tipo de usuário.
# Tenha um boxplots para cada ano e analise o que ocorreu
df_restringido = df_mesclado[df_mesclado['colecao'] == colecao][(df_mesclado['tipo_vinculo_usuario'] == 'ALUNO DE GRADUAÇÃO') | (df_mesclado['tipo_vinculo_usuario'] == 'ALUNO DE PÓS-GRADUAÇÃO')]
df_restringido = df_restringido[['id_emprestimo', 'ano_emprestimo', 'mes_emprestimo', 'tipo_vinculo_usuario', 'colecao']]
df_restringido
```

```python

meses = df_restringido['mes_emprestimo'].unique()
relacao_emprestimos_meses_graduacao = pd.DataFrame({})
relacao_emprestimos_meses_pos = pd.DataFrame({})

for mes in meses:
    df_restringido_graduacao = df_restringido[df_restringido['tipo_vinculo_usuario'] == 'ALUNO DE GRADUAÇÃO']
    df_restringido_pos = df_restringido[df_restringido['tipo_vinculo_usuario'] == 'ALUNO DE PÓS-GRADUAÇÃO']
    relacao_emprestimos_meses_graduacao[str(mes)] = [ len(df_restringido_graduacao[df_restringido_graduacao['mes_emprestimo'] == mes]) ]
    relacao_emprestimos_meses_pos[str(mes)] = [ len(df_restringido_pos[df_restringido_pos['mes_emprestimo'] == mes]) ]
relacao_emprestimos_meses_graduacao
```

```python
sns.barplot(data=relacao_emprestimos_meses_graduacao)
```

```python

meses = df_restringido['mes_emprestimo'].unique()
anos = df_restringido['ano_emprestimo'].unique()
relacao_emprestimos_meses_graduacao_por_ano = pd.DataFrame({})
relacao_emprestimos_meses_pos_por_ano = pd.DataFrame({})

for ano in anos:
    meses_grad = pd.DataFrame({})
    meses_pos = pd.DataFrame({})
    for mes in meses:
        df_restringido_graduacao = df_restringido[df_restringido['tipo_vinculo_usuario'] == 'ALUNO DE GRADUAÇÃO']
        df_restringido_pos = df_restringido[df_restringido['tipo_vinculo_usuario'] == 'ALUNO DE PÓS-GRADUAÇÃO']
        
        meses_grad[str(mes)] = [ len(df_restringido_graduacao[df_restringido_graduacao['mes_emprestimo']==mes][df_restringido_graduacao['ano_emprestimo']==ano]) ]
        meses_pos[str(mes)] = [ len(df_restringido_pos[df_restringido_pos['mes_emprestimo']==mes][df_restringido_pos['ano_emprestimo']==ano]) ]

    relacao_emprestimos_meses_graduacao_por_ano = pd.concat([relacao_emprestimos_meses_graduacao_por_ano, meses_grad], ignore_index=True)
    relacao_emprestimos_meses_pos_por_ano = pd.concat([relacao_emprestimos_meses_pos_por_ano, meses_pos], ignore_index=True)
```

```python
meses = df_restringido['mes_emprestimo'].unique() # já ordenado
meses_extenso = ['Janeiro', 'Fevereiro', 'Março', 'Abril', 'Maio', 'Junho', 'Julho', 'Agosto', 'Setembro', 'Outubro', 'Novembro', 'Dezembro']
relacao_emprestimos_meses_graduacao_por_ano = relacao_emprestimos_meses_graduacao_por_ano.rename(columns=lambda x: meses_extenso[int(x)-1])
relacao_emprestimos_meses_pos_por_ano = relacao_emprestimos_meses_pos_por_ano.rename(columns=lambda x: meses_extenso[int(x)-1])
```

```python
anos = df_restringido['ano_emprestimo'].unique() # também já ordenado
relacao_emprestimos_meses_graduacao_por_ano = relacao_emprestimos_meses_graduacao_por_ano.rename(index=lambda x: anos[x])
relacao_emprestimos_meses_pos_por_ano = relacao_emprestimos_meses_pos_por_ano.rename(index=lambda x: anos[x])
```

```python

plt.figure(figsize=(20, 8))
relacao_emprestimos_meses_graduacao_por_ano.transpose().plot(kind='bar', stacked=True)
plt.xlabel('Mês')
plt.ylabel('Quantidade de Empréstimos')
plt.title('Distribuição de Empréstimos Mensais por Ano (Alunos de Graduação)')
plt.legend(title='Ano')
plt.xticks(rotation=45)
plt.show()
```
